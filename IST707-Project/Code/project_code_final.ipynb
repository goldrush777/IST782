```python
#Load data

import pandas as pd

#Loading Data
cbb_data = pd.read_csv("/workspaces/707-group-project-ncaa-seeds/data/cbb_ref_all.csv")

kenpom1523 = pd.read_csv("/workspaces/707-group-project-ncaa-seeds/data/kenpom1523.csv")
kenpom24 = pd.read_csv("/workspaces/707-group-project-ncaa-seeds/data/kenpom24.csv")
kenpom = pd.concat([kenpom1523, kenpom24], ignore_index=True)

conf_data = pd.read_csv("/workspaces/707-group-project-ncaa-seeds/data/conference_strength.csv")

#Merging Data
cbb_data2 = pd.merge(cbb_data, kenpom, left_on=["Season", "School"], right_on=["Season", "TeamName"], how="left")
cbb_full = pd.merge(cbb_data2, conf_data, left_on=["Season", "Conf"], right_on=["Season", "Conference"], how="left")

pd.set_option('display.max_columns', None)
cbb_full.head()
```

```python
cbb_full['seed'] = cbb_full['seed'].fillna(17)
```

```python
cbb_full['Wpct_Conf'] = cbb_full['Conf_W'] / (cbb_full['Conf_W']+cbb_full['Conf_L'])
cbb_full['Wpct_Home'] = cbb_full['Home_W'] / (cbb_full['Home_W']+cbb_full['Home_L'])
cbb_full['Wpct_Away'] = cbb_full['Conf_W'] / (cbb_full['Away_W']+cbb_full['Away_L'])
cbb_full.head()
```

```python
columns_to_remove = ['PF', 'Opp_PF','Rk', 'G', 'W', 'L', 'MP', 'FG', '3P', 'Conf_W', 'Conf_L', 'Home_W', 'Home_L', 'Away_W', 'Away_L',
                     'FT', 'Opp_MP', 'RankTempo', 'RankOE', 'RankAdjOE', 'RankAdjDE','RankDE','RankAdjEM', 'RankAdjTempo',
                     'Opp_FG', 'Opp_3P', 'Opp_FT', 'Conf', 'SOS.1', 'SRS.1', 'ORtg.1', 'TeamName', 'Rank']

cbb_full2 = cbb_full.drop(columns=columns_to_remove)
cbb_full2.head()
```

```python
cbb_full3 = cbb_full2.dropna()
cbb_full3.shape

teams = pd.array(cbb_full3.loc[cbb_full3['Season'] == 2024, 'School'])
confs = pd.array(cbb_full3.loc[cbb_full3['Season'] == 2024, 'Conference'])

cbb_full4 = cbb_full3.drop(columns = ['School','Conference'])
```

### Preprocessing

```python
cbb_z = cbb_full4.copy()

columns_to_exclude = ['Season', 'Wpct', 'seed', 'Rating', 'Wpct_Conf', 'Wpct_Home', 'Wpct_Away'] 

def standardize_column(column):
    return (column - column.mean()) / column.std()

columns_to_standardize = cbb_z.columns.difference(columns_to_exclude).tolist()

# Group by 'Season' and standardize the specified columns
cbb_z[columns_to_standardize] = cbb_z.groupby('Season')[columns_to_standardize].transform(standardize_column)


cbb_z.head(5)
```

```python
cbb_z.shape
```

```python
cbb_z['seed'] = cbb_z['seed'].astype(int)

test_data = cbb_z[cbb_z['Season'] == 2024]
train_data = cbb_z[cbb_z['Season'] != 2024]

test_data = test_data.drop(columns=['Season'])
train_data = train_data.drop(columns=['Season'])

test_data.reset_index(drop=True, inplace=True)
train_data.reset_index(drop=True, inplace=True)

X_train = train_data.drop(columns=['seed'])
X_train = X_train.astype(float)
y_train = train_data['seed']
X_test = test_data.drop(columns=['seed'])
X_test = X_test.astype(float)
y_test = test_data['seed']

X_train.head()
```

```python
X_train.shape
```

### Modeling

```python
import numpy as np
from sklearn.metrics import mean_squared_error

def bracket_predict(probs):
    weighted_average_prediction = np.dot(probs, np.arange(1, probs.shape[1] + 1))
    predictions = pd.DataFrame({ 'team':teams,'conf':confs,'seed': y_test, 'w_pred':weighted_average_prediction})
    min_values = predictions.groupby('conf')['w_pred'].min()

    predictions['status'] = 'no'

    for conf, min_value in min_values.items():
        conf_rows = predictions[predictions['conf'] == conf]
        conf_rows_sorted = conf_rows.sort_values(by='w_pred')
        min_row_index = conf_rows_sorted.index[0]
        predictions.loc[min_row_index, 'status'] = 'aq'
    
    aqs = predictions[predictions['status'] == 'aq']
    the_rest = predictions[predictions['status'] == 'no']
    at_larges = the_rest.sort_values(by='w_pred')

    at_larges = at_larges.head(36)

    the_field = pd.concat([aqs, at_larges]).sort_values(by='w_pred')
    the_field['prediction'] = np.repeat(range(1, 17), [4, 4, 4, 4, 4, 4, 4, 4, 4, 4, 6, 4, 4, 4, 4, 6])
    the_field.tail(60)

    result = pd.merge(predictions, the_field, on=['team','conf','seed','w_pred','status'], how='left')
    result['prediction'] = result['prediction'].fillna(17)

    hits = result[(result['seed'] < 17) & (result['prediction'] < 17)]

    percentage = (len(hits) / len(result[result['seed'] < 17])) * 100

    mse = mean_squared_error(hits['seed'], hits['prediction'])

    return result, the_field, percentage, mse
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

max_depth_values = [5, 10, 15]
min_samples_leaf_values = [1, 3, 5]
max_features_values = ['sqrt', 'log2']

best_params = {}

# Triple nested for loops
for max_depth in max_depth_values:
    for min_samples_leaf in min_samples_leaf_values:
        for max_features in max_features_values:
            # Create a RandomForestClassifier with current parameter values
            rf_classifier = RandomForestClassifier(
                n_estimators=100, 
                max_depth=max_depth, 
                min_samples_leaf=min_samples_leaf, 
                max_features=max_features, 
                random_state=42
            )
            
            # Perform cross-validation
            grid_search = GridSearchCV(rf_classifier, param_grid={}, cv=5)
            grid_search.fit(X_train, y_train)

            probabilities = grid_search.predict_proba(X_test)
            result, the_field, percentage, mse = bracket_predict(probabilities)
            print(max_depth, min_samples_leaf, max_features, percentage, mse)


```
